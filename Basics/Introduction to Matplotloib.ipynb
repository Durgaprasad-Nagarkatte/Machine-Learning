### Introduction to matplotlib

1) Using xlabel, ylabel, title, grid() function.  
2) Plotting discrete plots using 'ro'.  
3) Multiple plots.  

```python
import matplotlib.pyplot as plt
```

```python
# Plotting with just the Y-axis values
plt.plot([2, 4, 6, 4], linestyle='solid', marker='o', color='blue')
plt.ylabel("Numbers") 
plt.xlabel("Indices")
plt.title("MyPlot")
plt.show()
```

```python
# Plotting with X-Axis and Y-Axis
plt.plot([1, 2, 3, 4], [1, 4, 9, 16], '--bo')
plt.ylabel('squares')
plt.xlabel('numbers')
plt.title('Squares of numbers')
plt.grid()
plt.axis([0, 5, 0, 18])
plt.show()
```

```python
# Discrete Plot
plt.plot([1, 2, 3, 4], [1, 8, 27, 64], 'ro')
plt.title('Cubes of numbers')
plt.xlabel('numbers')
plt.ylabel('cubes')
plt.grid()
```

```python
# Plot with multiple values
import numpy as np
t = np.arange(0, 5, 0.2)

plt.plot(t, t**2, 'b--', label='^2')
plt.plot(t, t**2.2, 'rs', label='^2.2')
plt.plot(t, t**2.5, 'g^', label='^2.5')
plt.grid()
plt.legend()
plt.show()
```

```python
# A plot with thick linewidth
x = [1, 2, 3, 4]
y = [1, 4, 9, 16]
plt.plot(x, y, linewidth=5.0)
plt.show()
```

```python
# Multiple plot
x1 = [1, 2, 3, 4]
y1 = [1, 4, 9, 16]
y2 = [2, 4, 6, 8]

lines = plt.plot(x1, y1, x1, y2)
plt.setp(lines[0], color='r', linewidth=1.5, label='squares')
plt.setp(lines[1], color='g', label='doubles')
plt.legend()
plt.grid()
```

**Working with mutiple plots**

```python
def f(t):
    return np.exp(-t) * np.cos(2*np.pi*t)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.4)

# Creates a figure
plt.figure(1)

# Creates 2 suplots, 2 rows, 1 column, and 1st plot.  
plt.subplot(211)
plt.grid()
plt.plot(t1, f(t1))
plt.title('Figure - 0.1')


plt.subplot(212)
plt.grid()
plt.plot(t2, f(t2))
plt.title('Figure - 0.4')
```

```python
plt.figure(1)
plt.grid()

plt.subplot(211)
plt.plot([1, 2, 3], 'ro')

plt.subplot(212)
plt.plot([4, 5, 6], 'go')

plt.figure(2)
plt.plot([4, 5, 7])

plt.figure(1)
plt.subplot(211)
plt.title('Numbers')
```

```python
data = {'a': np.arange(50),
        'c': np.random.randint(0, 50, 50),
        'd': np.random.randn(50)}
data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 100

plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()
```
