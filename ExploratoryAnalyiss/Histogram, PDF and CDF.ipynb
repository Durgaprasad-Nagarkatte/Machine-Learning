```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv("data/iris.csv", names=columns)
iris['species'].value_counts()
```

```python
virginica = iris[iris["species"] == 'Iris-virginica']
setosa = iris[iris["species"] == 'Iris-setosa']
versicolor = iris[iris["species"] == 'Iris-versicolor']
```

## Histogram Plots and PDF function

```python
sns.FacetGrid(iris, hue="species", height=10)\
.map(sns.distplot, "petal_length", bins=10)\
.add_legend();
```

The smooth line is the **probability distribution function**.

## Cumulative Density Function

```python
counts, bin_edges = np.histogram(setosa['petal_length'], bins=10, density=True)
pdf = counts/(sum(counts))
print(pdf)
print(bin_edges)

cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:], pdf)
plt.plot(bin_edges[1:], cdf)
plt.xlabel("petal length")
plt.ylabel("probability")
```

**82 percent of setosa flowers have petal length less than 1.6**

### Calculating cumulative density function

```python
# number of flowers having petal length less than 1.6
count = len(setosa['petal_length'][setosa['petal_length'] < 1.6])
total = len(setosa)
cdf = count/total
print(cdf)
```
