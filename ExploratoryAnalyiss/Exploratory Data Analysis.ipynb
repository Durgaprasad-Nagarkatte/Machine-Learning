# Exploratory Data Analysis
**Things learnt**  
1) Plotting a scatter plot using matplotlib and seaborn  
2) Pair Plots (Visualizing higher dimensional data)  
3) Disadvantages of Pair Plots.  

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
iris = pd.read_csv("data/iris.csv", names=columns)
```

```python
# Number of data points in the dataset
print(iris.shape)
```

```python
# Columns in the dataset
print(iris.columns)
```

```python
iris["species"].value_counts()
```

## 2-D scatter plot

```python
iris.plot(kind='scatter', x='sepal_length', y='sepal_width')
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue='species', height=6)\
.map(plt.scatter, "sepal_length", "sepal_width")\
.add_legend()
plt.show()
```

```python
sns.set_style("whitegrid")
sns.FacetGrid(iris, hue='species', height=6)\
.map(plt.scatter, "petal_length", "petal_width")\
.add_legend()
plt.show()
```

**Use plotly to visualize data in 3D space** 

### Visualizing higher dimensional data

```python
plt.close()
sns.set_style("whitegrid")
sns.pairplot(iris, hue="species", height=3)
plt.show()
```

By observing the above data we can say that petal length and petal width distinguishes the 3 different species of flowers.

## Limitation of pair plot

Pair plots have a major disadvantage as it works only for low dimensional data. Consider a dataset having 10 features, then the number of plots would be 45 which is difficult to analyse. But there are dimensionality reduction techniques like PCA, t-SNE which helps visualizing data in lower dimension.
