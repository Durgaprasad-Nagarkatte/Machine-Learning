```python
import numpy as np
import seaborn as sns
from scipy import stats 
import matplotlib.pyplot as plt
```

```python
# Generating a Gaussian Random Variable
x = stats.norm.rvs(size=1000)
sns.set_style('whitegrid')
sns.kdeplot(np.array(x), bw=0.5)
plt.show()
```

```python
stats.kstest(x, 'norm')
```

#### Since the p-value is high we can say that the x is normally distributed.

```python
# Uniform Distribution
y = np.random.uniform(0, 1, 10000)
sns.kdeplot(np.array(y), bw=0.1)
plt.show()
```

```python
stats.kstest(y, 'norm')
```

#### As p-value is zero the random variable is not normally distributed
